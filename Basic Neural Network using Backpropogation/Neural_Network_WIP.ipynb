```python
def sigmoid(x):
    stored_sigmoid = []
    for i in x:
        stored_sigmoid.append(1./(1.+ np.exp(-x[i])))
    return sum(stored_sigmoid)
```

```python
def neuron(inputs, Hweights, Oweights, Hbias, Obias):
    stored_hidden = []
    for i in inputs:
        stored_hidden.append(np.dot(inputs, Hweights[:,i]) + Hbias[i])
    stored_output = []
    for j in stored_hidden:
        stored_output.append(np.dot(stored_hidden, Oweights[:,j]) + Obias[j])
    stored_output = np.asarray(stored_output)
    return sigmoid(stored_output)
```
